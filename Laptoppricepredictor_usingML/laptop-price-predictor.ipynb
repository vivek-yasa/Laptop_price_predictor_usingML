Z-Score Method

```python
from scipy.stats import zscore

# Compute Z-scores for each column
z_scores = np.abs(zscore(df[['Inches', 'Ram', 'Weight', 'Price']]))

# Set a threshold (e.g., 3) for identifying outliers
threshold = 1

# Create a boolean mask for outliers
outlier_mask = z_scores > threshold

# Remove rows containing outliers
df_cleaned = df[~outlier_mask.any(axis=1)]
print("Z-Scores:\n", z_scores)
print("Outlier Mask:\n", outlier_mask)
```

Interquartile Range (IQR) Method

```python
Q1 = df[['Inches', 'Ram', 'Weight', 'Price']].quantile(0.25)
Q3 = df[['Inches', 'Ram', 'Weight', 'Price']].quantile(0.75)
IQR = Q3 - Q1

# Create a boolean mask for outliers
outlier_mask = ((df[['Inches', 'Ram', 'Weight', 'Price']] < (Q1 - 1.5 * IQR)) | (df[['Inches', 'Ram', 'Weight', 'Price']] > (Q3 + 1.5 * IQR)))

# Remove rows containing outliers
df_cleaned = df[~outlier_mask.any(axis=1)]
```

```python
1. **Standard Error (SE)**:
   - Use when you want to show the precision of the sample mean as an estimate of the population mean.
   - Appropriate when you are dealing with sample means and want to convey how representative the sample means are
    of the true population mean.
   - Often used in inferential statistics to display the variability in sample means.

2. **Standard Deviation (SD)**:
   - Use when you want to show the spread or dispersion of individual data points within each category.
   - Provides information about how data points are scattered around the mean.
   - Useful when you want to emphasize the variation in the data.

3. **Interquartile Range (IQR)**:
   - Use when you want to show the spread of the middle 50% of the data, which is less affected by extreme values (outliers).
   - Useful for non-normal distributions or when you want to emphasize the central tendency of the data.

4. **Minimum and Maximum Values**:
   - Use when you want to show the range of data points within each category.
   - Useful for displaying the overall spread of the data.

5. **Custom Error Values**:
   - Use when you have specific error values calculated or estimated based on your data or experimental design.
   - Allows you to convey domain-specific information about uncertainty.

To choose the appropriate measure of variability, consider the following:
- The characteristics of your data: Is it normally distributed? Are there outliers?
- The goals of your analysis: Do you want to emphasize the central tendency, spread, or both?
- The message you want to convey: What is the key point you want your audience to take away from the visualization?

```

```python
# Define a function to categorize processor types
def fetch_processor(text):
    # Check if the processor type is one of the known Intel Core types
    if text in ['Intel Core i7', 'Intel Core i5', 'Intel Core i3']:
        return text  # Return the same processor type if it matches known Intel Core types
        
    else:
        # Check if the processor type starts with 'Intel'
        if text.split()[0] == 'Intel':
            return 'Other Intel Processor'  # Categorize as 'Other Intel Processor'
        else:
            return 'AMD Processor'  # If not an Intel processor, categorize as 'AMD Processor'
```

```python
# Convert 'Memory' column values to strings and remove decimal '.0'
df['Memory'] = df['Memory'].astype(str).replace('\.0', '', regex=True)

# Remove 'GB' and replace 'TB' with '000' in the 'Memory' column
df["Memory"] = df["Memory"].str.replace('GB', '')
df["Memory"] = df["Memory"].str.replace('TB', '000')

# Split 'Memory' column into two parts based on '+'
new = df["Memory"].str.split("+", n=1, expand=True)

# Create 'first' and 'second' columns from the split
df["first"] = new[0]
df["first"] = df["first"].str.strip()
df["second"] = new[1]

# Create columns for different types of storage in 'first'
df["Layer1HDD"] = df["first"].apply(lambda x: 1 if "HDD" in x else 0)
df["Layer1SSD"] = df["first"].apply(lambda x: 1 if "SSD" in x else 0)
df["Layer1Hybrid"] = df["first"].apply(lambda x: 1 if "Hybrid" in x else 0)
df["Layer1Flash_Storage"] = df["first"].apply(lambda x: 1 if "Flash Storage" in x else 0)

# Remove non-numeric characters from 'first' column
df['first'] = df['first'].str.replace(r'\D', '')

# Fill missing values in 'second' column with "0"
df["second"].fillna("0", inplace=True)

# Create columns for different types of storage in 'second'
df["Layer2HDD"] = df["second"].apply(lambda x: 1 if "HDD" in x else 0)
df["Layer2SSD"] = df["second"].apply(lambda x: 1 if "SSD" in x else 0)
df["Layer2Hybrid"] = df["second"].apply(lambda x: 1 if "Hybrid" in x else 0)
df["Layer2Flash_Storage"] = df["second"].apply(lambda x: 1 if "Flash Storage" in x else 0)

# Remove non-numeric characters from 'second' column
df['second'] = df['second'].str.replace(r'\D', '')

# Convert 'first' and 'second' columns to integers
df["first"] = df["first"].astype(int)
df["second"] = df["second"].astype(int)

# Calculate storage quantities based on storage types and layers
df["HDD"] = (df["first"] * df["Layer1HDD"] + df["second"] * df["Layer2HDD"])
df["SSD"] = (df["first"] * df["Layer1SSD"] + df["second"] * df["Layer2SSD"])
df["Hybrid"] = (df["first"] * df["Layer1Hybrid"] + df["second"] * df["Layer2Hybrid"])
df["Flash_Storage"] = (df["first"] * df["Layer1Flash_Storage"] + df["second"] * df["Layer2Flash_Storage"])

# Drop unnecessary columns
df.drop(columns=['first', 'second', 'Layer1HDD', 'Layer1SSD', 'Layer1Hybrid',
                 'Layer1Flash_Storage', 'Layer2HDD', 'Layer2SSD', 'Layer2Hybrid',
                 'Layer2Flash_Storage'], inplace=True)
```

```python
# Define a function named 'cat_os' that categorizes operating systems into broader groups.
# The function takes an input 'inp', which represents an operating system.

def cat_os(inp):
    # Check if the input operating system is Windows 10, Windows 7, or Windows 10 S.
    if inp == 'Windows 10' or inp == 'Windows 7' or inp == 'Windows 10 S':
        return 'Windows'  # Return the category 'Windows' for these operating systems.
    # Check if the input operating system is macOS or Mac OS X.
    elif inp == 'macOS' or inp == 'Mac OS X':
        return 'Mac'  # Return the category 'Mac' for these operating systems.
    else:
        return 'Others/No OS/Linux'  # Return the category 'Others/No OS/Linux' for other cases.
```

```python
import numpy as np
import pandas as pd

# Define a function to remove outliers using IQR method
def remove_outliers_iqr(df, column, lower_bound=0.25, upper_bound=0.75):
    q1 = df[column].quantile(lower_bound)
    q3 = df[column].quantile(upper_bound)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Specify columns for which you want to remove outliers
columns_to_remove_outliers = ['Ram', 'Weight', 'Price', 'ppi', 'HDD', 'SSD']

# Remove outliers for each specified column
for column in columns_to_remove_outliers:
    df = remove_outliers_iqr(df, column)

# Display the modified DataFrame without outliers
print(df)
```

### Linear regression

```python
step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop='first'),[0,1,7,10,11])
],remainder='passthrough')

step2 = LinearRegression()

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```

### Ridge Regression

```python
step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop='first'),[0,1,7,10,11])
],remainder='passthrough')

step2 = Ridge(alpha=10)

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```

```python
from sklearn.ensemble import VotingRegressor, StackingRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, ExtraTreesRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score, mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder

# Assuming you have X_train, X_test, y_train, y_test already defined

# Define the ColumnTransformer
step1 = ColumnTransformer(transformers=[
    ('col_tnf', OneHotEncoder(sparse=False, drop='first'), [0, 1, 7, 10, 11])
], remainder='passthrough')

# Define the individual estimators
rf = RandomForestRegressor(n_estimators=350, random_state=3, max_samples=0.5, max_features=0.75, max_depth=15)
gbdt = GradientBoostingRegressor(n_estimators=100, max_features=0.5)
xgb = XGBRegressor(n_estimators=25, learning_rate=0.3, max_depth=5)
et = ExtraTreesRegressor(n_estimators=100, random_state=3, bootstrap=True, max_samples=0.5, max_features=0.75, max_depth=10)

# Define the VotingRegressor with weights
step2 = VotingRegressor([('rf', rf), ('gbdt', gbdt), ('xgb', xgb), ('et', et)], weights=[5, 1, 1, 1])

# Create the pipeline
pipe = Pipeline([
    ('step1', step1),
    ('step2', step2)
])

# Fit the pipeline and make predictions
pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)

# Calculate and print R-squared score and MAE
print('R2 score:', r2_score(y_test, y_pred))
print('MAE:', mean_absolute_error(y_test, y_pred))
```
